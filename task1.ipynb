```python
import numpy as np
import matplotlib.pyplot as plot

number_of_dots = 100
X = 10 * np.random.rand(number_of_dots)
y = (np.random.rand() - 0.5) * 5 * X + np.random.rand(number_of_dots) * 10

sum_all_x_and_y = np.sum(X * y)
sum_all_x = np.sum(X)
sum_all_y = np.sum(y)
sum_all_x_sqr = np.sum(X**2)

k = (sum_all_x_and_y - sum_all_x * sum_all_y / number_of_dots) / (
    sum_all_x_sqr - sum_all_x**2 / number_of_dots
)
b = np.mean(y) - k * np.mean(X)
y_computed = k * X + b

plot.title("Линейная регрессия")
plot.xlabel("x")
plot.ylabel("y")
plot.scatter(X, y, color="blue")
plot.plot(X, y_computed, color="red", linewidth=3)
plot.show()
```
